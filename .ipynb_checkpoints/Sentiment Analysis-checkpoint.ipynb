<a id="sa-introduzione"></a>
<h3>Introduzione</h3>

<p style="text-align: justify;">La Sentiment Analysis,  chiamata anche Opinion Mining, è il campo di studio il cui obbiettivo è quello di estrarre l’opinione e il sentimento da un testo scritto utilizzando metodi computazionali. In altri termini la SA mira a determinare l’atteggiamenti dell’autore dello scritto analizzato rispetto ad un determinato argomento o l’intero documento <a href="#sa-riferimenti-2">$[2]$</a>.<br>

Nel dizionario Treccani il sentimento viene definito come “la facoltà e l’atto del sentire, di avvertire impressioni esterne o interne” mentre l’opinione è definita come “un concetto che una persona si forma riguardo a particolari fatti o fenomeni”. La differenza tra i due termini è sottile: l’opinione concerne il proprio punto di vista riguardo a qualcosa mentre il sentimento è una sensazione. Tale sfumatura è irrilevante ai fini della SA, perciò utilizzeremo in maniera indistinta l’uno e l’altro termine.<br>
    
<p style="text-align: justify;">Il termine Sentiment Analysis viene incominciato ad usare solo agli inizi degli anni 2000, questo perché prima di questa data i testi scritti e contenenti un’opinione in forma digitale erano praticamente inesistenti. Con l’esplosione del web e dei social media, questi sono cresciuti esponenzialmente. Ora ognuno ha la possibilità di esprimere liberamente le proprie opinioni e punti di vista circa qualsiasi argomento e in ogni angolo del mondo, grazie alle nuove piattaforme quali Twitter, Facebook, Instagram e Tripadvisor, Youtube, solo per citarne alcune. A partire dal 2002 si rilevano infatti già numerose ricerche nel campo della SA. Il mondo ha cominciato a capire quanto le opinioni e i sentimenti abbiano un enorme potere, in quanto centrali per gran parte delle attività umane. Gli esseri umani spesso nel prendere decisioni, ricercano l’opinione altrui. E questo vale non solo per gli individui, ma anche per le aziende.<br>

Infatti, nonostante la SA derivi dal campo dell’informatica, viene oggi utilizzata in numerosi e disparati ambiti quali il management, la politica, l’economia e in tutti quei campi che hanno a che fare con i consumatori e l’opinione pubblica. 

<p>Normalmente la SA viene applicata su tre diversi livelli:
<ul>
    <li>
        <p style="text-align:justify;">A livello di Documento: l’obbiettivo è determinare se l’intero documento esprime un sentimento positivo, negativo o neutro. A questo livello di analisi si assume che ogni documento esprime un’opinione relativamente a una singola entità (intesa come un singolo prodotto o servizio). Ed è quindi non applicabile a documenti che valutano o comparano più entità.
    </li>
        <br>
    <li>
        <p>A livello di Frase: qui lo scopo è classificare se ogni frase esprime un’opinione positiva, negativa o neutra. 
    </li>
        <br>
    <li>
        <p>A livello di Aspetto: invece che guardare alle unità di linguaggio (documenti, paragrafi, frasi) ha per oggetto le opinioni e il loro target (chiamato opinion target), cioè l’entità a cui si riferiscono. Per esempio, se noi della frase “Mi piace la tua auto” sappiamo solo che è positiva ma non a quale entità l’opinione positiva si riferisce, questo non c’è molto utile. Si prenda ad esempio la frase “Nonostante il servizio non fosse buono, amo ancora quel ristorante”, questa ha un tono positivo, ma non possiamo affermare che l’intera frase è positiva. L’affermazione è positiva relativamente al ristorante non rispetto al servizio. Con l’opinion targets il ristorante è considerato come un’entità e il servizio come uno degli aspetti dell’entità. L’obbiettivo di questo livello di analisi è quello di determinare il sentimento nei confronti delle entità e/o dei loro aspetti. 
Questo livello di analisi è quello che viene utilizzato maggiormente.
    </li>
        <br>
</ul>
<p>I testi oggetto di analisi della SA possono essere classificati in due diverse categorie:
<ol>
    <li>
        <p>Factual based (objective): espressioni oggettive relative ad entità, eventi e i loro attributi, come per esempio “Ieri sono stato in montagna”.
    </li>
        <br>
    <li>
        <p>Opinion based (subjective): espressioni soggettive che esprimono sentimenti, attitudini, emozioni o sensazioni attraverso entità, eventi e i loro attributi, come ad esempio “ Mi piace veramente tanto andare in montagna”. 
    </li>
        <br>
</ol>
<p>Normalmente la SA performa meglio sui testi che hanno un contenuto soggettivo rispetto a quelli con un contenuto solo oggettivo. Questo perché quando uno scritto è oggettivo, esso esprime fatti senza trasmettere nessuna emozione, sentimento o attitudine. Cosa che invece troviamo nei testi Opinion based, che sono generalmente social media, sondaggi, articoli e dati contenenti feedback. Detto ciò anche gli scritti Factual based possono implicare sentimenti positivi o negativi dell’autore perché possono descrivere fatti desiderabili o indesiderabili e sono quindi oggetto della SA. Si pensi ad esempio alle espressioni oggettive come “Ieri ho comprato un paio di scarpe e oggi si sono rotte” oppure “Dopo un mese che hanno costruito la casa, si è formata una crepa nel muro”, esse descrivono due fatti indesiderati e possiamo con sicurezza affermare che l’autore delle affermazioni si esprime negativamente rispetto alle scarpe e alla casa.
<p>Un’ulteriore differenza viene individuata sul tipo di opinioni oggetto della SA, che possono essere classificate in :
<ul>
    <li>
        <p>Regular opinions: esprimono un sentimento riguardo una precisa entità o un aspetto dell’entità, per esempio “La pasta al pesto è veramente molto buona” esprime un’opinione o un sentimento positivo sul gusto della pasta al pesto. Questo rappresenta il tipo di opinione più comune.
    </li>
        <br>
    <li>
        <p>Comparative opinions: mettono a confronto più entità sulla base di alcune delle loro caratteristiche comuni. Per esempio “La pasta al pesto è più buona di quella al ragù” compara la pasta al pesto e al ragù sulla base del loro gusto (una delle caratteristiche) ed esprime una preferenza per la prima.
    </li>
        <br>
</ul>

<a id="sa-applicazioni"></a>
<h3>Applicazioni</h3>
<ul>
    <li>
        <p>Social Media Monitoring: le aziende sono sempre molto interessate nel conoscere le opinioni e i giudizi dei clienti rispetto ai propri prodotti e servizi.  Oggi queste non sono più tenute a condurre questionari e sondaggi per reperire queste informazioni, ma possono farlo monitorando la propria opinione e interagendo con i consumatori attraverso i social network quali Facebook, Twitter e Instagram.  La SA consente di analizzare velocemente i feedback degli acquirenti, classificarli secondo il livello di urgenza e rispondere quindi attivamente a richieste o critiche. 
    </li>
        <br>
    <li>
        <p>Brand Monitoring: le società possono reperire le informazioni non solo sui social media ma anche in internet. Attraverso l’analisi di notizie, blog e forum sono in grado di comprendere la propria brand reputation.
    </li>
        <br>
    <li>
        <p>Ricerca di mercato: la SA potenzia e rafforza le ricerche di mercato e di analisi competitiva. Sia se si vuole esplorare nuovi mercati, anticipare i trends furi o fare un’analisi dei propri competitor. Lo studio delle proprie recensioni rispetto a quelle dei concorrenti, l’ analisi dei report di mercato e dei business journals per identificare i trends del futuro e l’analisi dei tweets e dei social media in tempo reale sono solo alcuni dei possibili strumenti a supporto della propria strategia.
    </li>
        <br>
    <li>
        <p>Lo stato e i governi possono ottenere facilmente dei feedback dell’opinione pubblica concernenti le politiche esistenti o di quelle proposte monitorando i propri social media. 
    </li>
        <br>
    <li>
        <p>Gli stessi consumatori prima di acquistare determinati prodotti e servizi ricercano giudizi di chi ne ha già fatto esperienza, i quali sono facilmente accessibili pubblicamente. 
    </li>
        <br>
    <li>
        <p>Negli ultimi anni la SA ha trovato applicazione praticamente in ogni ambito: dal turismo all’ospitalità, dai servizi in ambito finanziario agli eventi sociali, dalla politica all’ assistenza sanitaria. 
    </li>
</ul>
<p>Il fatto che più grandi aziende quali Google, Microsoft, Amazon, eBay, SAS, Oracle, Adobe e Bloomberg si sono dotate di propri sistemi interni di SA ci fa capire quanto la SA sia un importante strumento a servizio della strategia aziendale.

<a id="sa-definizione-formale"></a>
<h3>Definizione Formale</h3>

<p>Date le premesse delle sezioni precedenti è opportuno dare una descrizione più formale e approfondita del problema della SA concentrandoci sulla definizione dell’opinione e sulle sue caratteristiche.<br>
L’opinione è un concetto ampio che comprende il sentimento, la valutazione, l’attitudine, l’entità e/o gli aspetti dell’entità oggetto di opinione e il soggetto che la esprime. <br>
L’opinione ha due componenti chiave: un target $g$ e un sentiment $s$ riferito al target, $(g, s)$, in cui $g$ è una qualsiasi entità ($e$) o aspetto ($a$) dell’entità su cui è stata espressa l’opinione.<br>
Dove l’entità può essere un prodotto, un servizio, una persona, un evento o un’organizzazione e l’aspetto è una delle caratteristiche o attributi di $g$. Prendendo ad esempio una macchina fotografica come entità, questa è formata da una serie di componenti quali obbiettivo, batterie, schermo ed ha una serie di caratteristiche come risoluzione, pixel, prezzo. $S$ invece è il sentimento, questo può essere positivo, negativo neutro o espresso in scala numerica.<br>
Ulteriori componenti dell’opinione sono l’opinion holder e il time of opinion ovvero rispettivamente il soggetto che esprime l’opinione e il momento in cui questa è stata espressa.

<p>Riassumendo l’opinione è rappresentata dalla tupla $(e, a, s, h, t)$, dove:
    
<ul>
    <li>
        <p>$e$ è l’entità su cui l’opinione viene espressa
    </li>
    <li>
        <p>$a$ è uno degli aspetti dell’entità
    </li>
    <li>
        <p>$s$ il sentimento dell’opinione formulata sull’aspetto a dell’entità e
    </li>
    <li>
        <p>$h$ è il soggetto che ha pronunciato l’opinione
    </li>
    <li>
        <p>$t$ il tempo in cui l’opinione è stata espressa.
    </li>
</ul>

<p>Nei casi in cui l’opinione si riferisce unicamente all’entità per intero, viene utilizzato l’aspetto $\text{GENERAL}$.<br>
In questo caso $e$ ed $a$ insieme rappresentano l’opinion target.

<p>Prendiamo ad esempio la recensione:

<div style="border-style:solid;border-width:2px;padding:10px;margin-top:20px;">
<p><i>(1)I bought a Canon G12 camera six months ago. (2) I Simply love it. (3) The picture quality is amazing. (4) The battery life is also long. (5) However, my wife thinks it is too heavy for her.</i><br><br>
    <span>
        <b style="text-align:left;">Review A</b> : Posted by Jhon Smith 
    </span>
<br>
    <span>
        <b style="text-align:right;">Date</b> : September 10, 2011
    </span> 
</div>


<p>Nella recensione A la fotocamera Canon G12 rappresenta l’entità $e$, mentre la picture quality, la battery life e il weight rappresentano gli attributi $a$ dell’entità Canon G12.<br>
Insieme formano il target del sentimento $s$, il quale è positivo nelle affermazioni 2, 3 e 4 e negativo nella 5.<br>
Gli opinion holders $h$ sono due: l’autore della recensione “John Smith” per le frasi dall’ 1 alla 4 e la moglie dell’autore frase 5.<br>
Infine il tempo $t$ in cui la recensione è stata formulata è indicato dalla data “September 10, 2011”.

<p>Un documento $d$ contiene un’ opinione relativamente ad un set di entità $\{e_1, e_2,\cdots , e_n\}$ e  un sotto insieme di aspetti $a = \{a_1, a_2, \cdots , a_n\}$  per ogni entità $e_i$.<br>
Le opinioni sono espresse da un insieme di opinion holders $\{h_1, h_2,\cdots ,h_n\}$ e vengono espresse in un determinato istante temporale $t$.<br>
Dato un insieme di documenti $D$, compito della sentiment analysis è:<br>

<ol>
    <li>
        <p>Entity extraction and resolution: estrarre tutte le espressioni contenenti un’entità in $D$ e dividerli in categorie.
    </li>
<br>
    <li>
        <p>Aspect extraction and resolution: estrarre tutte le espressioni contenenti gli aspetti di una data entità e dividerli in categorie. Dove ogni categoria rappresenta un singolo aspetto $a$.
    </li>
<br>
    <li>
        <p>Opinion holder extraction and resolution: per ogni opinione identificata nel testo estrarre il soggetto che l’ha espressa.
    </li>
<br>
    <li>
        <p>Time extraction and standardization: individuare il momento in cui ogni opinione viene espressa.
    </li>
<br>
    <li>
        <p>Aspect sentiment classification or regression: determinare se un opinion riguardo un aspetto a o un’entità e è positiva, negativa o neutrale o assegnarli un punteggio.
    </li>
<br>
    <li>
        <p>Opinion quintuple generation: per ogni opinione individuata in $D$ produrre la tupla $(e, a, s, h, t)$ sulla base dei risultati ottenuti nei punti 1-5.
    </li>
</ol>

<p>Vediamo di seguito un esempio pratico del processo sovradescritto applicato ad una recensione :<br>
    
<div style="border-style:solid;border-width:2px;padding:10px;margin-top:20px;">
<p><i>(1)I bought a Samsung camera and my friend bought a Canon camera yesterday. (2) In the past week we both used the cameras a lot. (3) The photos from my Samy are not clear for night shots, and the battery life is short too. (4) My friend was very happy with his camera and loves its picture quality. (5) I want a camera that can take good photos. (6) I am going to return it tomorrow.</i><br><br>
    <span>
        <b style="text-align:left;">Review B</b> : Posted by bigJhon
    </span>
<br>
    <span>
        <b style="text-align:right;">Date</b> : September 15, 2011
    </span> 
</div>

<p>Il punto 1 estrae le entità <b>Samsung</b>, <b>Samy</b> e <b>Canon</b> e raggruppa Samsung e Samy insieme perché rappresentano la stessa entità.<br>
Il punto 2 estrae gli aspetti <b>picture, photo e battery life</b> e raggruppa picture e photo insieme dal momento che sono sinonimi.<br>
Il punto 3 identifica come opinion holder <b>bigJohn</b> (nonché autore della recensione) relativamente a tutte le affermazioni contenute nella recensione fatta eccezione per la numero 4 in cui l’opinion holder è <b>l’amico di bigJohn</b>.<br>
Il punto 4 poi individua <b>“September 15, 2011”</b> come tempo in cui le opinioni sono state espresse.<br>
Il punto 5 stabilisce che la frase (3) da un’opinione negativa rispetto alla picture quality e alla battery life della fotocamera Samsung, mentre l’espressione (4) ne da una positiva riguardo la fotocamera Canon come entità e alla sua picture quality.<br>
Infine il punto 6 genera, per ogni opinione individuata, un riassunto delle sue caratteristiche:

<ol>
    <li>$$(\text{Samsung},\,\text{picture_quality},\,\text{negative},\,\text{bigJhon},\,\text{Sep. 15, 2011})$$
    </li>
<br>    
    <li>$$(\text{Samsung},\,\text{battery_life},\,\text{negative},\,\text{bigJhon},\,\text{Sep. 15, 2011})$$
    </li>
<br>    
    <li>$$(\text{Canon},\,\text{GENERAL},\,\text{positive},\,\text{bigJhons_friend},\,\text{Sep. 15, 2011})$$
    </li>
<br>    
    <li>$$(\text{Canon},\,\text{picture_quality},\,\text{positive},\,\text{bigJhons_friend},\,\text{Sep. 15, 2011})$$
    </li>
</ol>

<p>Nella maggior parte delle applicazioni è necessario analizzare opinioni che provengono da più opinion holders, è quindi opportuno farne un riassunto. Una dei modelli più utilizzati è chiamato aspect-based opinion summary o feature based-opinion summary. Data un’entità $e$ l’<b>Aspect-based opinion summary</b> si struttura in questa forma:<br><br>

<table style="width:100%;">
    <tr>
        <th style="text-align:center">Aspect</th>
        <th style="text-align:center">$|OH^+|$</th>
        <th style="text-align:center">$|OH^-|$</th>
    </tr>
    <tr>
        <td style="text-align:center">$\text{GENERAL}$</td>
        <td style="text-align:center">$x_{1,1}$</td>
        <td style="text-align:center">$x_{1,2}$</td>
    </tr>
    <tr>
        <td style="text-align:center">$\text{Aspect}_1$</td>
        <td style="text-align:center">$x_{2,1}$</td>
        <td style="text-align:center">$x_{2,1}$</td>
    </tr>
    <tr>
        <td style="text-align:center">$\vdots$</td>

        <td style="text-align:center">$\vdots$</td>
    </tr>
    <tr>

<p>Dove $|OH^+|$ e $|OH^-|$ rappresentano rispettivamente la cardinalità dell'insieme di opinion holders che hanno espresso un giudizio positivo o negativo nei confronti di un dato aspetto dell'entità in questione.<br>
$\text{GENERAL}$ rappresenta l’entità stessa $e$, valutata nel suo complesso, mentre $\text{Aspect}\,n$ rappresenta l'aspetto $n-\text{esimo}$ dell'entità $e$.
L’elemento di forza di questo riassunto è il fatto che metta in evidenza il numero di opinioni positive e negative rispetto ad ogni entità e ai suoi aspetti.

<p>E' opportuno sottolineare come le definizioni e i concetti introdotti sopra, nonostante siano stati presentati nel contesto di recensioni su un prodotto, abbiano valenza generale e siano applicabili a tutte le forme di testi che esprimono una qualche opinione. Ad esempio potrebbe trattarsi di news, tweets, discussioni su forum, blogs e/o post Facebook.
    
<p>Chiaramente diversi domini di applicazione comportano differenti livelli di difficoltà. È più agevole trattare opinioni relative a prodotti e servizi rispetto a quelle riguardanti temi sociali e politici, non solo per la complessità dell’argomento ma anche per l’utilizzo di espressioni sarcastiche e ironiche.

<a id="sa-processo"></a>
<h3>Il Processo della Sentiment Analysis</h3>
<p>La SA è un processo complesso che si può suddividere in quattro steps <a href="#sa-riferimenti-3">$[3]$</a>:<br>
<ol>
    <li><h5>Data Collection</h5>
        <p>Il primo passo consiste nel raccogliere i dati, a seconda del nostro obbiettivo, da bolgs, social networks, forums, articoli di giornali, recensioni. Questi si presentano come disorganizzati, infatti il linguaggio umano non è uniforme, ma al contrario si manifesta in modi diversi tramite l’utilizzo di diversi registri linguistici, vocabolari, slangs,  etc. Un’analisi di tipo manuale su questi dati sarebbe impossibile. Pertanto occorre passare alla fase successiva.
    </li>
    <li><h5>Text Preprocessing</h5>
        <p>Questo step consiste nel pulire il testo estratto e prepararlo affinchè su di esso sia possibile condurre l’analisi del sentiment. I contenuti non testuali e quelli irrilevanti ai fini dell’analisi vengono quindi identificati e poi eliminati.
    </li>
    <li><h5>Sentiment Classification</h5>
        <p>In questa fase le parole, le frasi o l’intero documento vengono classificate come positive, negative o neutre. In seguito approfondiremo i principali approcci attraverso cui la classification del sentiment può essere eseguita.
    </li>
    <li><h5>Results Analysis</h5>
        <p>Una volta conclusa l’analisi, i risultati ottenuti vengono visualizzati tramite grafici come grafici a torta, istogrammi etc.
    </li>
</ol>

<a id="sa-processo-dataCollection"></a>
<h4>Data Collection</h4>

<p>Come abbiamo precedentemente osservato l'idea alla base della SA è l'estrapolazione di un sentiment da dati testuali.

<p>Un requisito fondamentale per il processo di SA è dunque la disponibilità di dati. Le fonti d'informazione sono molteplici; ad oggi le persone condividono conoscienza, esperienze e opinioni, principalmente, attraverso social media quali blogs, forums, siti di recensioni, social networks, e così via <a href="#sa-riferimenti-3">$[3]$</a>.

<p>I metodi di acquisizione dei dati variano in funzione delle fonti e delle interfaccie utilizzate per la loro condivisione. Una prima distinzione può essere effettuata sulla base dell'accessibilità dei dati, siano essi di aziende private o pubblicamente disponibili.

<h5>I dati nelle aziende</h5>

<p>Ad oggi le aziende riconoscono la vitale importanza della gestione dei dati generati dai loro servizi. Per questo motivo le informazioni riguardo ai propri clienti sono spesso raccolte in basi di dati relazionali. Un aspetto altrettano importante, oltre alla raccolta dati, è la loro protezione. E' infatti molto difficile riuscire ad ottenerne l'accesso ed è dunque necessario rivolgersi a fonti di dati pubblicamente accessibili.

<h5>Dati free-source</h5>

<p>Tra le fonti di dati pubblicamente disponibili troviamo : <br>

<ul>
    <li>Data sets pubblici, disponibili ad esempio sulla piattaforma <a href="https://archive.ics.uci.edu/ml/index.php">UCI Irvine</a>.</li><br>
    <li>Fonti governative ufficiali, quale <a href="https://www.dati.gov.it/">dati.gov.it</a></li>
<br>
    <li>Dati privati esposti selettivamete tramite interfaccie, ad esempio le API di Twitter, Facebook, etc. <br>
   Le API rappresentano dei punti di accesso a funzionalità per l'implementazione di programmi che facilitano il riuso del codice, snelliscono i processi di sviluppo software e migliorano l'interoperabilità tra sistemi.<br>
    Per quello che riguarda il nostro caso di studio, un esempio di API è l'interfaccia esposta da Twitter. Questa interfaccia può essere impiegata per accedere a dati ed informazioni relative agli utenti della piattaforma ma permettono anche di interagire con la piattaforma ad esempio pubblicando, leggendo o modificando tweets.<br>
        Un esempio di utilizzo delle API di Twitter è presente nel <a href="./Twitter API.ipynb">relativo notebook</a>.
    </li>
<br>
    <li>Dati ottenuti tramite web scraping da web crawlers e bots.<br>
    Il web costituisce una fonte di informazioni non strutturate molto ricca, qualora i dati non siano esposti in maniera programmatica è quindi necessario implementare un meccanismo di raccolta. Una tecnica ampliamente utilizzata a questo fine sfrutta i <i>web crawlers</i> o i <i>bots</i>, programmi istruiti nel navigare il web analizzando le pagine e raccogliendo le informazioni richieste.</li>
</ul>

<p>Per l'interoperabilità tra sistemi e per facilitare la condivisione sono stati definiti formati standard per la condivisione dei dati. In merito alcuni dei formati ad oggi più utilizzati sono : csv, xml e json.
I file csv sono semplici file di testo in cui ad ogni riga corrisponde un dato le cui features sono separate da virgola. <br>

<p>Il formato XML, dall'inglese eXtensible Markup Language, permette la definizione di linguaggi utili alla descrizione di documenti strutturati. E' infatti possibile, tramite tags, definire degli schemi che descrivono la sintassi e la semantica dei documenti. Questo formato, assieme a json, costituisce uno dei principali meccanismi di scambio dati sul Internet.

<a id="sa-processo-textPreprocessing"></a>
<h4>Text Preprocessing</h4>
<p>Una volta ottenuto il testo che si vuole analizzare, questo non è ancora pronto per essere esaminato. Oggi infatti, più dell’80% dei dati sono unstructured e non già accuratamente ordinati in tabelle e schemi. Inoltre, il linguaggio umano possiede delle strutture molto complesse che è necessario decodificare e interpretare. Per esempio i testi ottenuti dai social network, essendo scritti in un linguaggio spesso informale, sono debolmente strutturati : presentano di frequente errori di battitura, di ortografia, URLs, tag, hastagh, slangs etc.
Al fine di trarre da questi dati informazioni utili il primo passo è il pre-processing del testo. Questo è uno step fondamentale della SA che consiste in una serie di fasi ci permettono di trasformare il testo nella forma più congeniale affinchè gli algoritmi di ML possano performare al meglio su di essi. In particolare è opportuno cercare di ridurre il testo, che rappresenta un dato di tipo qualitativo, a uno quantitativo, in modo tale che esso possa rappresentare l’input del modello di ML con il quale si vuole svolgere l’analisi del sentiment. Tale processo viene definito Feature Extraction e consiste appunto nel trasformare i dati, come testi ed immagini in features numeriche che possono essere interpretate dagli algortimi di ML <a href="#sa-riferimenti-4">$[4]$</a>.<br>
Numerosi paper in questo campo hanno dimostrato l’utilità e la necessità del pre-processing del testo al fine della feature extraction. Per citarne alcuni <a href="#sa-riferimenti-5">$[5]$</a> prima e <a href="#sa-riferimenti-6">$[6]$</a> successivamente, hanno dimostrato come, attraverso l’utilizzo di alcune tecniche di pre-processing applicate su dei tweets, queste abbiano facilitato e migliorato l’accuratezza nella classificazione del sentiment. Anche <a href="#sa-riferimenti-7">$[7]$</a> e <a href="#sa-riferimenti-8">$[8]$</a> sono arrivati a dimostrare lo stesso esaminando in questo caso recensioni di film online.<br>
I principali benefici del pre-processing del testo sono:
<ul>
    <li>Limitare il fenomeno dell’overfitting, che si verifica quando un modello statistico complesso si adatta eccessivamente ai dati osservati perché ha un numero eccessivo di parametri <a href="#sa-riferimenti-9">$[9]$</a>. Ridurre la complessità del testo ci permette infatti di ottenere data sets con un minor numero di features, facilmente interpretabili da algoritmi, limitandone il numero di parametri e quindi arrivare ad ottenere un modello in grado di generalizzare.
</li><br>
    <li>Ridurre la dimensione dei file, Kannan e Gurusamy hanno rilevato come lo step della rimozione delle 'stopwords' e dello 'stemming', che verranno introdotti in seguito, arrivano a ridurre rispettivamente del 25% e 45% il numero degli indici nel testo <a href="#sa-riferimenti-10">$[10]$</a>. 
</li>
</ul>
<p>Il pre-processing può consistere in diversi steps a seconda dei dati che si hanno a disposizione, del contesto e della situazione. L’approccio utilizzato sarà quindi diverso a seconda che si tratti di recensioni o di un articolo piuttosto che dei tweets o dei commenti. 
In generale però ritroviamo spesso le seguenti fasi :
<ul>
    <li><b>Tokenizazione :</b><br> 
    <p>Per analizzare un testo dal punto di vista computazionale è necessario insegnare al computer a riconoscere gli elementi di cui questo è composto. Il processo preliminare a qualsiasi tipo di elaborazione del testo è la sua tokenizzazione. La Tokenization ci consente di dividere la string di testo in unità minime di analisi, chiamate tokens (parole, punteggiatura, numeri, sigle ecc.).
    </li><br>
    <li><b>Normalizzazione :</b><br> 
    <p>La normalizzazione è il processo di trasformazione del testo in una forma standard, in modo da ridurre l’ammontare di informazioni che il computer dovrà in seguito analizzare. Essa può consistere ad esempio nel trasformare un numero nell’equivalente forma scritta, nel convertire l’intero testo in carattere minuscoli, tramutare le emoticons nell’emozione corrispondente etc. Questo passaggio è molto importante quando si esaminano testi quali commenti nei social media, recensioni di prodotti o servizi, tweets etc, in quanto in essi sono spesso presenti abbreviazioni, errori di battitura e ortografici e slangs.<br>
        Non c’è un procedimento standard per normalizzare il testo, questo dipende infatti dall’obbiettivo che abbiamo. Un'operazione di particolare importanza durante la fase di normalizzazione è la rimozione delle stopwords.<br>
    Le stopwords sono le parole più comuni di un testo, quelle che non hanno a che vedere con un argomento in particolare, non sono rilevanti nel contesto dei dati e non contribuiscono a chiarire il significato della frase. Gli articoli, le proposizoni e le congiunzioni sono classici esempi di stopword. In questa fase le stopwords vengono eliminate in quanto occupano molte risorse (spazio di memoria, tempo di ricerca) e non contengono alcun sentimento. Inoltre può accadere che la presenza delle stopwords porti fuori strada l’algoritmo perché crea collegamenti semantici dove non ci sono. Al fine di eliminare le stopwords è sufficiente confrontare ogni parola del testo che si intende analizzare con un elenco di stopwords.
    </li><br>
    <li><b>Stemming &amp; Lemmatization :</b><br> 
    <p>La lingua che parliamo e scriviamo è infatti formata da parole che spesso derivano da altre. In questo caso le parole si dicono in forma flessa e alcune di queste condividono una radice comune.<br>
        <img src="./imgs/sa_roots.png" style="width:400px">
    <p>Lo stemming e la Lemmatization ci aiutano entrambi a ottenere la radice delle parole flesse, nel farlo utilizzano però due approcci diversi.Lo stemming è il processo di riduzione della forma flessa di una parola alla sua forma radice “stem”<a href="#sa-riferimenti-11">$[11]$</a>, anche se lo stem di per sé non sempre rappresenta una parola corretta/valida nella lingua (trouble $\rightarrow$ troubl).  Nello stemming la radice è un elemento irriducibile della parola che continua ad esprimere il significato del termine complesso, cioè del termine in forma flessa.<br>
        <img src="./imgs/sa_stemming.png" style="width:350px">
    <p>In generale lo stemming tende a eliminare i suffissi come -ed, -ing, -er, i plurali e le forme possessive. Nel testo da analizzare quindi si sostituiscono tutte le parole con le loro radici. Ciò che si ottiene è un testo con la stessa quantità di termini ma con meno varianti, cosa che rende più efficace il successivo processo di analisi. Ancora una volta è importante sottolineare come con lo stemming di una parola o di una frase non sempre si ottengano parole che sono valide nella lingua.<br>
    Questo processo viene implementato tramite l’uso di un algoritmo. Il più utilizzato per lo stemming in lingua inglese è l’algoritmo di Porter (1980), il quale consiste in 5 fasi sequenziali, in ognuna delle quali la parola viene ridotta.
    <img src="./imgs/sa_stemming_ex1.png" style="width:350px">
    <p>Lo stemming è efficace a patto che il processo di riduzione della parola a radice non faccia perdere informazioni e la radice della parola conservi lo stesso significato della parola completa <a href="#sa-riferimenti-12">$[12]$</a>.<br>
    Per superare i limiti dello stemming sono stati sviluppati degli algoritmi che utilizzano il lemma delle parole al posto della radice. Dove il lemma rappresenta la forma canonica o del dizionario di una serie di parole. Ad esempio il lemma delle forme verbali ‘are’, ‘is’, ‘being’ è il verbo all’infinito ‘be’. La Lemmatization, a differenza dello Stemming, riduce le parole in forma flessa assicurandosi che il lemma appartenga al linguaggio. Per utilizzare la lemmatization bisogna quindi tenere in considerazione l’analisi morfologica della parola, per fare questo è necessario avere un dizionario dettagliato a cui l’algoritmo può fare riferimento per collegare le parola che analizza al lemma corrispondente.<br><br>

        <p>Stemming o lemmatization?<br>
        Entrambe generano la radice della parola flessa. La differenza è che lo stem non sempre rappresenta una parola valida nel linguaggio, mentre il lemma lo è.
        Inoltre lo stemming utilizza un algoritmo a più step che performa su ogni parola, cosa che lo rende più veloce ed efficiente. Nella lemmatization invece dobbiamo consultare per ogni parola il database WordNet e definire a parts of speech per ottenere il lemma corretto, cosa che lo rende più lento e meno efficiente dello stemming.<br>
        Quindi l’utilizzo di una tecnica piuttosto che di un’altra dipende dal testo su cui stiamo lavorando: se abbiamo bisogno di velocità ed efficienza oppure di un output le cui parole siano valide nel linguaggio.

<a id="sa-processo-sentimentClassification"></a>

<h4>Sentiment Classification</h4>

<p>Rappresenta la terza fase della SA e  concerne la classificazione delle parole, frasi o dell’intero documento come positivo, negativo o neutro. Gli approcci utilizzati nella Sentiment Classification sono principalmente tre:<br>

<img src="./imgs/sa_sentiment_classification.png" style="width:500px;"><br>

<ul>
    <li><b>Approcio Lexicon based :</b><br>

<p>Le parole, i modi di dire e le frasi che all’interno di un testo trasmettono sentimenti positivi o negativi sono strumenti chiave per la SA. Queste vengono chiamate Sentiment Lexicon.  Ad esempio le parola 'divertente', 'bello', 'magnifico' trasmettono positività, mentre 'triste', 'brutto' emozioni negative. L’approccio Lexicon Based si basa sulla creazione di un lessico, cioè una lista di parole a ognuna delle quali viene associato un sentimento specifico. A differenza dell'approccio basato su ML, questo metodo non necessita di allenare il modello, la difficoltà principale sta nel creare un lessico univoco che potrà poi essere impiegato nella successiva fase di analisi in contesti diversi.<br>
Tre sono i principali metodi utilizzati per la costruzione di un lessico <a href="#sa-riferimenti-1">$[1]$</a>: l’<b>approccio manuale</b>, l’<b>approccio dictionary-based</b> e quello <b>corpus based</b>.

<p>L'<b>approccio manuale</b> consiste nel costruire il lessico manualmente assegnando un valore di polarità a un elenco scelto di parole. Esso richiede molto tempo e lavoro e viene normalmente utilizzato per testare gli approcci automatici e verificare se questi hanno fatto errori.

<p>Nell’<b>approccio dictionary based</b> il lessico viene costruito partendo da un piccolo set di parole (semi), raccolte manualmente e a cui viene assegnato un orientamento positivo o negativo. Questo set viene ampliato successivamente tramite un algoritmo che cerca nei dizionari sinonimi e contrari delle parole già nell’elenco. Le nuove parole trovate vengono aggiunte alla lista esistente. Questo processo termina quando non vengono trovate altre parole <a href="#sa-riferimenti-13">$[13]$</a>.<br>
Dopo avere completato questo processo normalmente si fa un controllo manuale per pulire la lista e rimuovere eventuali errori. Il vantaggio di questo metodo è che è possibile trovare facilmente e velocemente un grande numero di sentiment words e il loro orientamento (positivo o negativo). Il principale svantaggio è rappresentato dal carattere generale di questi lessici, la sentiment orientation delle parole ottenute in questo modo è indipendente dal contesto e del dominio e non sempre riesce a cogliere le differenze di orientamento delle parole utilizzate in contesti diversi. Ad esempio, dire che il microfono del computer è silenzioso indica generalmente un sentimento negativo. Mentre associare l’aggettivo silenzioso a un elettrodomestico suggerisce un orientamento positivo.

<p>Per cercare di risolvere questo problema viene utilizzato un <b>approccio corpus-based</b>. 
Mentre il precedente approcio prevede l'uso di dizionari al fine di determinare l'orientamento semantico delle parole, l'approcio corpus-based si affida a co-occrenze statistiche e pattern sintattici contenuti nei testi. L'intuizione sottostante questo approcio è che la distanza semantica tra una parola e l'insieme di semi, positivi e negativi, possa costituire una metrica per la stima della polarità del sentimento della parola target in questione <a href="#sa-riferimenti-18">$[18]$</a>. Questo approcio permette, oltre alla creazione di lessici ad-hoc per un determinato dominio, anche l'adattamento di lessici 'general-pourpose' ad domini d'applicazione specifici.<br>
I metodi per la costruzione di lessici basati su testi variano e le tecniche adoperate in letteratura sono molteplici. Di particolare interesse sono però gli approci basati su propagazione delle labels e quelli basati su informazione mutua puntuale o PMI, dall'inglese pointwise mutual information. I primi sfruttano strutture quali grafi pesati, costruiti sulla base di un insieme di seeds e successivamente estesi, per inferire la polarità di parole nuove. I secondi utilizzano invece il concetto di informazione mutua, una misura che quantifica la dipendenza di due variabili casuali, per quantificare la similitudine in termini di sentiment tra parole nuove e parole appartenenti all'insieme di seeds.

<p>Tramite l’utilizzo di uno dei metodi sopra illustrati abbiamo ottenuto un lessico nel quale ad ogni parola è stato associato una certa polarità, o positiva o negativa. Nella maggior parte dei casi il calcolo della polarità totale del testo viene eseguito calcolando la media delle polarità delle singole parole tramite <b>algoritmi rules-based</b>. L’approccio rules based utilizza un set di regole create dagli umani volte ad aiutare a identificare la soggettività, la polarità o il soggetto di un’opinione. Un esempio base di come un sistema rule-based lavora è il seguente: viene creato un lessico all’interno del quale a ciascuna parola viene assegnata una valenza positiva o negativa. Si contano poi il numero di parole negative e positive all’interno del testo, se il numero di parola positive supera quello delle negative allora il sistema assegna al testo carattere positivo, altrimenti negativo. Se invece il risultato è zero, il testo viene considerato neutrale. Naturalmente quello presentato è un sistema molto semplice e naive. Esistono tecniche basate su regole molto più complesse e nuove regole possono sempre essere aggiunte al fine di meglio cogliere il vocabolario e le espressioni presenti nel testo. In generale viene impiegata la formula :<br><br>

$$\frac{\sum_{w \in t}{S_{w}\cdot \text{Weight}(w)\,\cdot \text{Modifier}(w)}}{\sum_{w \in t}{\text{Weight}(w)}}$$

<p>Dove $t$ indica il documento da analizzare, $w$ le parole di cui $t$ è composto mentre $S_w$ è la polarità associata al termine $w$. $\text{Weight}(w)$ e $\text{Modifier}(w)$ rappresentano invece due funzioni utilizzate per dare più o meno peso a determinate parole.<br>
Esistono molti tool che sfruttano l’approccio lexicon based e rule based.   SentiWordNet è un lessico molto utilizzato, il quale si basa sul dizionario WordNet. WordNet è un database semantico-lessicale della lingua inglese, elaborato presso l’università di Princeton nel 1985 che si propone di organizzare, definire e descrivere i concetti espressi dai vocaboli. Questo è costituito da synset, ovvero raggruppamenti di termini con significato affine. La maggior parte dei synet è poi connessa ad altri synet da relazioni semantiche <a href="#sa-riferimenti-14">[14]</a>.<br>
SentiWordNet <a href="#sa-riferimenti-15">[15]</a> è un lessico nel quale ogni synset di WordNet è associato a tre punteggi numerici $\text{Obj}(s)$, $\text{Pos}(s)$, $\text{Neg}(s)$, i quali descrivono quanto oggettivo, positivo o negativo è il termine contenuto nel synet <a href="#sa-riferimenti-16">[16]</a>.<br>
Tali punteggi sono compresi tra i valori 0 e 1 e, messi assieme, forniscono la polarità del termine in questione e possono essere rappresentati graficamente all’interno di un triangolo come in figura $7$.

<img src="./imgs/sa_word_polarity.png" style="width:350px">

<p style="font-size:12px;text-align:right;"><i>Immagine $7$, cortesia di <a href="http://ontotext.fbk.eu/sentiwn.html">SentiWordNet</a></i><br>

<p>$\text{Pos}(w)$ e $\text{Neg}(w)$ indicano rispettivamente il punteggio di positività e negatività assegnato al termine in questione. Mentre il livello di oggettività/soggettività viene ricavato dalla relazione : <br><br>

$$\text{Obj} = 1 – (\text{Pos}(w) + \text{Neg}(w))$$

<p>dove $\text{Obj} = 0$ corrisponde al massimo grado di oggettività mentre $\text{Obj} = 1$ il massimo livello di soggettività.

    <li>La capacità di lavorare in maniera efficace anche avendo pochi dati a disposizione.</li>

    <li>E’ stato dimostrato ampiamente in letteratura <b style="color:red">[Cit. needed]</b> come, una volta scelto il lessico corretto in rapporto al problema da trattare, gli approci basati su lessico forniscano grande accuratezza.</li>
</ul>

<p>Sfortunatamente vi sono però anche svantaggi nell'impiegare questi approci :<br>

<ul>
    <li>
        Come già in precedenza abbiamo sottolineato, l'orientamento di alcune parole è fortemente dipendente dal contesto e dal dominio di applicazione.
    </li>
<br>
    <li>Nonostante una parola possa comparire nel lessico creato con un particolare sentiment, non sempre è detto che tale parola venga utilizzata per esprimere un sentimento. Questo accade spesso nelle frasi interrogative e condizionali. Consideriamo ad esempio le frasi : "Mi potresti dire se domani il tempo è bello?" e “Se oggi al supermercato trovi della bella frutta compramela”. Nonostante entrambe contengano la parola 'bello', nessuna delle due esprime un sentimento, positivo o negativo, nè nei confronti del tempo nè rispetto alla frutta. Tuttavia questo non implica che nessuna frase condizionale o interrogativa non possa trasmettere un sentimenti, come gli esempi "Di chi è questa borsa orribile?" e "Se non hai voglia di andare a quello squallido pub rimani a casa" dimostrano.</li>
<br>
    <li>Con gli approci basati su lessico inoltre risulta difficile trattare le frasi sarcastiche come “E’ proprio un bravo politico, ha distrutto un paese”. Oppure frasi come “che bella fregatura!” nella quale compaiono sia “bella” che “fregatura”, le quali assumono valenze rispettivamente positive e negative, facendo risultare il valore finale della frase neutrale, anche quando risulta evidente il giudizio negativo che esprimono.</li>
<br>
    <li>Molte frasi, pur non contenendo nessuna parola ritenuta portatrice di un sentimento, possono implicare un’ opinione positiva o negativa del loro autore. Per esempio “Questa macchina consuma molta benzina” implica un’opinione negativa dell’auto perché consuma molte risorse ed è quindi dispendiosa.</li>
<br>
    <li>Infine, risulta difficile creare un dizionario ontologico che sia costantemente aggiornato, soprattutto quando si analizzano testi informali come quelli di social media, blogs e forums.
</ul>
<br>
<li><b>Approci basati su machine learning:</b><br>

<p>Gli approcci basati su machine learning puntano alla creazione di modelli in grado di apprendere e migliorare autonomamente sulla base dei dati a nostra disposizione. Il prossimo capitolo svilupperà in maniera più approfondita questo argomento, per ora ci limitiamo a darne una breve introduzione.<br>

<p>Una prima classe di tecniche di machine learning, detta supervised learning (SL), utilizza un insieme di dati già etichettati, detto insieme di training, per addestrare il modello che successivamente verrà utilizzato per la fase di predizione.
L'idea alla base del SL è l'esistenza di una funzione $f$ che modelli il fenomeno che si stia cercando di predirre mettendo in relazione i dati in input e quelli in output. Il nostro modello può dunque essere rappresentato come un'approssimazione parametrizzata di tale funzione. Il processo di traning mira dunque all'aggiustare tali parametri in modo da minimizzare le discrepanze tra la funzione sottostante il fenomeno e la sua approssimazione.
Riassumendo i punti salienti di questo approccio possono essere così sintetizzati: si hanno delle variabili di input $x$ e una variabile di output $y$, si utilizza un algoritmo per definire una funzione che metta in relazione gli input agli output, $Y = \hat{f}(X)$. L’obbiettivo è quello di arrivare ad approssimare la funzione descrivente il fenomeno così bene che quando si hanno nuovi dati in input si è in grado di predirre correttamente gli output corrispondenti.
I requisiti fondamentali di questo sistema sono quindi la creazione del training set e la numerosità dello stesso. Se il training set dispone di un numero ridotto di dati è difficile che il computer riesca ad apprendere la corretta relazione tra input e output e sintetizzarla in una funzione $\hat{f}$ che approssimi f. Al contrario, se si dispone di un training set numeroso è più probabile che la funzione approssimi più accuratamente f e che quindi l’output elaborato dal modello sia corretto. <br>
Gli algoritmi maggiormente utilizzati sono Naive Bayes, Maximum Entropy (ME) e Support Vector Machines (SVM) <a href="#sa-riferimenti-17">$[17]$</a>.

<img src="./imgs/sa_ml_sentiment_analysis.png" style="width:600px">
    
<p>Il metodo Unsupervised viene impiegato quando non si dispone di testi già etichettati. Per l’analisi si dispone quindi solo degli input, non si conoscono gli output e non è quindi possibile modellare la funzione che mette in relazione tali entità.
L’obbiettivo dell’UL è quello di apprendere la struttura o le relazioni che intercorrono tra i  dati in modo da comprenderli meglio. Gli algoritmi sono lasciati liberi di scoprire pattern e strutture interessanti nei dati. Tra le tecniche di UL troviamo la Clustering Anlysis: si ha un problema di clustering quando quando si vuole raggruppare i dati in gruppi quanto più omogenei tra loro utilizzando il concetto di distanza. Una volta ottenuti i gruppi è necessario analizzare all’interno di ognuno le caratteristiche per le quali gli oggetti in quel gruppo sono simili e per quali invece si differenzino dagli altri gruppi.

<p>Il principale vantaggio degli approci basati su machine learning è la loro abilità nell’adattarsi e nel creare modelli per contesti e fini specifici. Mentre il  suo limite riguarda la necessità di tempo per allenare il modello e la disponibilità di una grande quantità di dati affinchè le predizioni siano sufficientemente affidabili.
