```python
import os
import re

import sys
sys.path.append("..")

import time
import pickle
import graphviz
import numpy as np
import pandas as pd
import wordcloud as wc
import matplotlib.pyplot as plt

from collections import defaultdict

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples

from scripts.plotting_functions import plot_inertia_silhouette, plot_clusters_wordclouds

from scripts.analyzers import ngrams_custom_tokenizer

from bokeh.plotting import output_notebook, show

output_notebook()


SEED = 99
np.random.seed(SEED)


DATA_FOLDER = os.path.join(os.getcwd(), "../data")
SETS_FOLDER = os.path.join(DATA_FOLDER, "csv")
```

```python
def clustering_results(X, n_clusters):

    results_df = pd.DataFrame(columns=['k', 'inertia', 'centroids', 'n_iter'])

    for cluster in n_clusters:
        kmean = KMeans(cluster)
        kmean.fit(X)
        inertia = kmean.inertia_
        centers = kmean.cluster_centers_
        labels = kmean.labels_
        sil_avg = silhouette_score(X, labels)
        n_iter = kmean.n_iter_
        partial_df = pd.DataFrame({
            'k':[cluster],
            'inertia':[inertia],
            'sil_avg':[sil_avg],
            'centroids':[centers],
            'n_iter':[n_iter]
        })
        results_df = results_df.append(partial_df)
        
    return results_df
```

```python
train_fp = os.path.join(SETS_FOLDER, "train.csv")
train_df = pd.read_csv(train_fp, index_col="review_id")
```

<h3>Clustering su mono e bi-grammi</h3>

```python
tfvc = TfidfVectorizer(
    lowercase=True, 
    tokenizer=ngrams_custom_tokenizer,
    ngram_range=(1,2)
)
```

```python
X = tfvc.fit_transform(train_df['text'])

n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10, 12, 14, 16, 18, 20]

bigrams_clusters_df = clustering_results(X, n_clusters)
```

```python
fig = plot_inertia_silhouette(bigrams_clusters_df, len(n_clusters))
show(fig)
```

<h3>Miglior numero di cluster</h3>

```python
kmeans = KMeans(6)
kmeans.fit(X)
```

```python
cluster_centers = kmeans.cluster_centers_
```

```python
top_features = defaultdict(list)

for i, cluster in enumerate(cluster_centers):
    for idx in np.argsort(cluster)[-20:]:
        top_features[i].append((idx, cluster[idx]))
```

```python
features_names = defaultdict(list)

for cluster in top_features:
    for idx, val in top_features[cluster]:
        for word in tfvc.vocabulary_:
            if(tfvc.vocabulary_[word] == idx):
                features_names[cluster].append((word, val))
```

```python
f1, axs1 = plot_clusters_wordclouds(features_names)
```
